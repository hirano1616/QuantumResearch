```python
from math import sqrt, floor, ceil
import gurobipy as gp
import random
import csv
```

```python
def makeExample(n,m,l, A_number):
    random.seed(m+l) #シード決定
    weight_item = [random.randrange(1,101) for i in range(m)] # 各アイテム重さを定義
    #乱数のシード値保証
    A_size = A_number #制約密度 （ローを示す）
    A = [[0 for j in range(n+1)] for i in range(m)] #制約式係数
    c = [0 for i in range(n+1)] #集合のコストを示す．

    S = [[] for i in range(n+1)] #集合Sを示す
    check_item = [0 for i in range(m)] #一度でも集合に属したか見る
    # print(S)
    # first Step: 各集合にアイテムを1つ割り当てる
    for j in range(n):
        v = random.randrange(m)
        S[j].append(v)
        # print(f"j={j}, v={v}, S[{j}]={S[j]}")
        A[v][j] = 1
        check_item[v] = 1

    # print("second step")
    # Second Step: 未割り当てのアイテムを集合に振り分ける．
    for i in range(m):
        if check_item[i] == 0:
            w = random.randrange(n) 
            S[w].append(i)
            # print(f"w={w} S[{w}]={S[w]}")
            A[i][w] = 1

    # print(S)
    Sum_S = 0
    for i in range(n):
        Sum_S += len(S[i])
    # print(Sum_S)

    # Third Step: 平均的な密度がA_sizeになるようにランダム振り分け
    for k in range(A_size*n - Sum_S):
        v = random.randrange(m)
        w = random.randrange(n)
        S[w].append(v)
        A[v][w] = 1

    # コスト計算
    # c = [0 for i in range(n+1)] #集合のコストを示す．
    for j in range(n):
        # print(f"j={j}")
        for i in S[j]:
            # print(f"i={i}")
            alpha = random.uniform(0.9, 1.1) #アイテムの重要度
            c[j] += alpha * weight_item[i] #重要度＊物の重さを加えていく．
        c[j] = floor(c[j]) #整数値に丸める．


    for i in range(m):
        # S[n].append(i)
        A[i][n] = 1
        c[n] += 2*weight_item[i]
    

    # print("作成したよん")
    return A,c
```

```python
n=30
m_list=[i for i in range(3, 30, 3)]
l_list=10
A_size = 3

for i in range(len(m_list)):
    m=m_list[i]
    A_size = int(sqrt(m))
    for j in range(l_list):
        l=j
        try:
            f1 = open(f'./dataset/example{n}_{m}_{l}_{A_size}.csv', 'w')
            writer1 = csv.writer(f1)
            A, c = makeExample(n,m,l,A_size)
            tmp_list = [m,n]
            writer1.writerow(tmp_list)
            writer1.writerow(c)
            A_list = [[i for i in range(m) if A[i][j] == 1] for j in range(n+1)]
            # print(A_list)
            writer1.writerows(A_list)
            
        finally:
            f1.close()
                
```
